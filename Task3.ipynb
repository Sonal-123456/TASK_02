```python
#Importing useful libraries
import numpy as num
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
#importing datas
dataset = pd.read_csv("C:/Users/sonal/The Sparks Foundation/Task 3/SampleSuperstore.csv")
```

```python
#reading the top datas
dataset.head()
```

```python
#checking for null values in dataset
dataset.info()
```

```python
#finding description of dataset
dataset.describe()
```

```python
#finding unique data
dataset.nunique()
```

```python
#evaluating the correlation matrix
ds = dataset.corr()
sns.heatmap(ds, xticklabels=ds.columns, yticklabels=ds.columns, annot=True, 
            cmap=sns.diverging_palette(310, 270, as_cmap=True))
```

```python
#ploting relationships between categorical data like segment and profit , ship mode and profit , state and profit , etc.
sns.catplot(x="Segment",y="Profit",data=dataset,kind='bar',ci=97)
```

```python
sns.catplot(x="Ship Mode",y="Profit",data=dataset,kind='bar',ci=None)
```

```python
order = dataset.groupby(["State"])["Profit"].mean().sort_values().index
sns.catplot(y="State",x="Profit",data=dataset,kind='bar',ci=None,height=15,order=order)
```

## we observe states like Ohio, North Carolina, Illinois yeild average loss

```python
sns.catplot(x="Category",y="Profit",data=dataset,kind="bar",ci=None)
```

## we observe that there is significantly high profits in the 'Technology' category and lesser profit in 'Furtniture' category.

```python
#for different regions
sns.catplot(x="Category",y="Profit",data=dataset,kind='bar',ci=None,col="Region")
```

## we observe that Selling Furtinure in Central region is not profitable.

```python
#count plot
plt.subplots(figsize=(8,10))
order=dataset.groupby(['State'])["State"].count().sort_values().index
sns.countplot(y='State',data=dataset,order=order)
```

## we conclude
1. We find most of the customers in California, New York and Texas. Hence these states can be generate more profit.
2. States like Wyoming, North Dakota and West Virginia have little or no sales at all. Hence, these states need to be reviewed.
3. More profit is generated with the sales of Technology goods, hence Sales can be broadened on for this category. On the other hand, furniture sales have not generated significant profits.
4. Selling Furtinure in Central region is loss-making.
5. Selling Furtinure in Eastern region is least profitable among all other regions and categories.
6. Profits among different segment do not vary much.
